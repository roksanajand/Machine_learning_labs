```python
def uproszcz_typ(typ):
    if typ in [
        'Przedszkole',
        'Punkt przedszkolny',
        'Zespół wychowania przedszkolnego'
    ]:
        return 'przedszkolne'
    
    elif typ in[ 'Szkoła podstawowa','Poznańska szkoła chóralna']:
        return 'podstawowe'
    
    elif typ in [
        'Liceum ogólnokształcące',
        'Technikum',
        'Branżowa szkoła I stopnia',
        'Branżowa szkoła II stopnia',
        'Szkoła policealna',
        'Szkoła specjalna przysposabiająca do pracy',
        'Bednarska Szkoła Realna'
        
    ]:
        return 'ponadpodstawowe'
    
    elif typ in [
        'Liceum sztuk plastycznych',
        'Ogólnokształcąca szkoła muzyczna I stopnia',
        'Ogólnokształcąca szkoła muzyczna II stopnia',
        'Ogólnokształcąca szkoła baletowa'
    ]:
        return 'artystyczne'
    
    else:
        return 'inne'
df['Grupa Podmiotu'] = df['Typ Podmiotu'].apply(uproszcz_typ)

# Sprawdzenie efektu
print(df['Grupa Podmiotu'].value_counts())
```

```python
# Słownik mapowania grupy na ID
mapa_id = {
    'przedszkolne': 1,
    'podstawowe': 2,
    'ponadpodstawowe': 3,
    'artystyczne': 4,
    'inne': 5
}

# Nowa kolumna z przypisanym ID
df['idGrupaPodmiotu'] = df['Grupa Podmiotu'].map(mapa_id)

# Sprawdzenie efektu
print(df[['Typ Podmiotu', 'Grupa Podmiotu', 'idGrupaPodmiotu']].drop_duplicates())
```

```python
# Grupowanie po powiecie i grupie podmiotu (szkoły przedszkolne, podstawowe, itd.)
agg = df.groupby(['Powiat', 'Grupa Podmiotu']).agg({
    'Liczba uczniów pobyt legalny': 'sum',
    'Liczba oddziałów': 'sum',
}).reset_index()

dodatki = df.groupby(['Powiat', 'Grupa Podmiotu']).agg({
    'Województwo': 'first',
    'idTerytPowiat': 'first',
    'idTerytWojewodztwo': 'first',
    'idPublicznosc': 'first',
    'Publiczność': 'first'
}).reset_index()

# 3. Łączenie agregacji i dodatkowych kolumn
agg = pd.merge(agg, dodatki, on=['Powiat', 'Grupa Podmiotu'], how='left')
agg = agg[[
    'Województwo',
    'idTerytWojewodztwo',
    'Powiat',
    'idTerytPowiat',
    'Publiczność',
    'idPublicznosc',
    'Grupa Podmiotu',
    'Liczba oddziałów',
    'Liczba uczniów pobyt legalny'
]]
# 4. Podgląd efektu
agg.head()
```

```python
def dostosuj_klase(row):
    if row['Klasa'] in ['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX']:
        if row['Grupa Podmiotu'] == 'podstawowe':
            return f"{row['Klasa']} podstaw."
        elif row['Grupa Podmiotu'] == 'ponadpodstawowe':
            return f"{row['Klasa']} ponadpodst."
        elif row['Grupa Podmiotu'] == 'artystyczne':
            return f"{row['Klasa']} artyst."
    return row['Klasa']  # zostaw inne bez zmian (np. sem. I, Wstępna itd.)

# Zastosowanie
df['Klasa'] = df.apply(dostosuj_klase, axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Oblicz średnią uczniów na oddział i sumę oddziałów
agg['Uczniowie_na_oddział'] = agg['Liczba uczniów pobyt legalny'] / agg['Liczba oddziałów']

# Grupowanie: oblicz średnią i sumę oddziałów dla każdej grupy podmiotu
grupy = agg.groupby('Grupa Podmiotu').agg({
    'Uczniowie_na_oddział': 'mean',
    'Liczba oddziałów': 'sum'
}).reset_index()

# Wykres
plt.figure(figsize=(8, 5))
ax = sns.barplot(
    data=grupy,
    x='Grupa Podmiotu',
    y='Uczniowie_na_oddział'
)
plt.title('Średnia liczba uczniów na oddział wg grupy podmiotu')
plt.ylabel('Uczniowie / oddział')
plt.xlabel('Grupa Podmiotu')

# Dodaj liczby oddziałów jako etykiety nad słupkami
for i, row in enumerate(grupy.itertuples()):
    ax.text(i, row.Uczniowie_na_oddział + 0.05, f"{int(row._3)} oddz.", ha='center')

plt.tight_layout()
plt.show()
```

```python
# Zakodowanie one-hot kolumny idPublicznosc
# Mapa ID -> Opisowa nazwa
publicznosc_map = {
    1.0: 'publiczna',
    4.0: 'niepubliczna',
    5.0: 'niepubliczna_artystyczna'
}

# Zamiana ID na opisy
publicznosc_named = agg['idPublicznosc'].map(publicznosc_map)

# One-hot encoding z nazwami opisowymi
publicznosc_ohe = pd.get_dummies(publicznosc_named, prefix='publicznosc', dtype=int)

# Dołączenie do agg
agg = pd.concat([agg, publicznosc_ohe], axis=1)

agg.head()
```
